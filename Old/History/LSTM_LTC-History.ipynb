## Model Training

```python
hyperparams = []
batch = [32, 64, 128]
epoch = [100]
neuron = [50, 60, 100]
for j in batch:
    for k in epoch:
        for l in neuron:
            hyperparams.append((j,k,l))
hyperparams
```

```python
lstms = []
models = []
for batch, epoch, neuron in hyperparams:
    model, lstm = LSTMUnit.train_lstm(train_X, train_y, test_X, test_y, neuron, epoch, batch)
    lstms.append(lstm)
    models.append(model)
```

```python
import pickle
i=0
for m in models:
    test_x2 = test_X
    yhat = m.predict(test_x2)
    inv_yhat = Preprocessing.inverse_scaler(yhat, scaler)
    inv_y = Preprocessing.inverse_scaler(test_y.reshape(-1,1), scaler)
    print(hyperparams[i])
    print("Epoch: "+ str(lstms[i].params['epochs']))
    print("Neurons: "+str(m.layers[0].units))
    print(Evaluation.mape(inv_y,inv_yhat)[0])
    with open('LSTM_LTC'+str(hyperparams[i])+'.pkl', 'wb') as f:
        pickle.dump(lstms[i].history, f)
    i = i+1
```
