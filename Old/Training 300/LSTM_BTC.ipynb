```python
from Skripsi import Preprocessing
from Skripsi import Evaluation
from Skripsi import LSTMUnit
from pandas import read_csv
from pandas import DataFrame
import numpy as np
import pandas as pd
from openpyxl.workbook import Workbook
import pickle
```

```python
btc_dfd = pd.read_csv('Binance_BTCUSDT_1h.csv')
btc_dfd
```

```python
df = Preprocessing.feature_selection(btc_dfd)
df
```

```python
df_no_dup = Preprocessing.handle_duplicate(df)
df_no_dup
```

```python
miss = Preprocessing.handle_missing_value(df_no_dup)
miss
```

```python
x, scaler = Preprocessing.minmax_scale(df_no_dup)
x
```

```python
train, test = Preprocessing.splitting_data(x)
train
```

```python
test
```

```python
train_X, train_y = Preprocessing.create_dataset(train,300)
test_X, test_y = Preprocessing.create_dataset(test,300)
```

```python
print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)
```

## Model Training

```python
hyperparams = []
batch = [32, 64, 128]
epoch = [25, 50, 100]
neuron = [50, 60, 100]
for j in batch:
    for k in epoch:
        for l in neuron:
            hyperparams.append((j,k,l))
hyperparams
```

```python
hyperparam1 = hyperparams[:9]
hyperparam2 = hyperparams[9:18]
hyperparam3 = hyperparams[18:27]
hyperparam1
```

```python
hyperparam2
```

```python
hyperparam3
```

```python
lstms1 = []
models1 = []
for batch, epoch, neuron in hyperparam1:
    model, lstm = LSTMUnit.train_lstm(train_X, train_y, test_X, test_y, neuron, epoch, batch)
    lstms1.append(lstm)
    models1.append(model)
```

```python
i=0
wb1 = Workbook()
ws1 = wb1.active
for m in models1:
    # make a prediction
    test_x2 = test_X
    yhat = LSTMUnit.predict(test_x2, m)
    inv_yhat = Preprocessing.inverse_scaler(yhat, scaler)
    inv_y = Preprocessing.inverse_scaler(test_y.reshape(-1,1), scaler)
    print(hyperparam1[i])
    print("Epoch: "+ str(lstms1[i].params['epochs']))
    print("Neurons: "+str(m.layers[0].units))
    i = i+1
    ws1['A'+str(i)] = 'LSTM'
    ws1['B'+str(i)] = hyperparam1[i-1][0]
    ws1['C'+str(i)] = hyperparam1[i-1][1]
    ws1['D'+str(i)] = hyperparam1[i-1][2]
    print('RMSE')
    print(Evaluation.rmse(inv_y,inv_yhat)[0])
    ws1['E'+str(i)] = Evaluation.rmse(inv_y,inv_yhat)[0]
    print('MAE')
    print(Evaluation.mae(inv_y,inv_yhat)[0])
    ws1['F'+str(i)] = Evaluation.mae(inv_y,inv_yhat)[0]
    print('MAPE')
    print(Evaluation.mape(inv_y,inv_yhat)[0])
    ws1['G'+str(i)] = Evaluation.mape(inv_y,inv_yhat)[0]
    LSTMUnit.save_model(m, 0, 'BTC',hyperparam1[i-1])
    with open('LSTM_BTC'+str(hyperparam1[i-1])+'.pkl', 'wb') as f:
        pickle.dump(lstms1[i-1].history, f)
wb1.save('LSTM_BTC_result1.xlsx')
```
